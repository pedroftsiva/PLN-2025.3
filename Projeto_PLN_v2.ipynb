```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)
```

```python
from langchain_community.document_loaders import YoutubeLoader
```

```python
def gera_transcricao(url:str)->str:
  loader = YoutubeLoader.from_youtube_url(
    url,add_video_info=False,language=["pt"]


  )
  transcricao = loader.load()[0].page_content

  return transcricao
```

```python
def resume_video(url:str)->str:
  transcricao = gera_transcricao(url)

  prompt = f"""
      Resuma o conte√∫do da do texto e uma √∫nica frase em ingl√™s, clara e objetiva

      Texto:
      {transcricao}
  """

  resposta = llm.invoke(prompt)

  return resposta.content
```

```python
import yt_dlp
import os

def get_voice(url, start_time=0, duration=120):

    output_file = "audio_youtube"

    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': f"{output_file}.%(ext)s",
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'postprocessor_args': [
            '-ss', str(start_time),
            '-t', str(duration)
        ],
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

    # procura o arquivo gerado
    final_audio_file = None
    for file in os.listdir():
        if file.startswith(output_file) and file.endswith(".mp3"):
            final_audio_file = file
            break

    if not final_audio_file:
        raise FileNotFoundError("‚ùå Arquivo de √°udio n√£o encontrado ap√≥s download!")

    full_path = os.path.abspath(final_audio_file)
    print(f"üéß √Åudio extra√≠do ‚Üí {full_path}")
    return full_path
```

```python
url = "https://youtu.be/dRhugEFaej8?si=v2RSxSDNj44Vz_QI"

resposta = resume_video(url)
caminho_voz = get_voice(url)
```
