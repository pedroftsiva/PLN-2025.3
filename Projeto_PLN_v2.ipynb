```python
import yt_dlp
import os

def get_voice(url, start_time=0, duration=120):
    output_file = "audio_youtube"

    ydl_opts = {
        # Formato mais genÃ©rico possÃ­vel
        'format': 'bestaudio/best',

        # Fallback automÃ¡tico
        'format_sort': ['res:720', 'ext:mp4:m4a', 'codec:avc'],

        'outtmpl': f"{output_file}.%(ext)s",
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'ignoreerrors': True,
        'force_generic_extractor': False,
    }

    # Adicionar cookies se o arquivo existir
    if os.path.exists('cookies.txt'):
        ydl_opts['cookiefile'] = '/content/drive/MyDrive/PLN/cookies.txt'

    # Corte de Ã¡udio
    if start_time > 0 or duration > 0:
        ydl_opts['postprocessor_args'] = ['-ss', str(start_time), '-t', str(duration)]

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        # Primeiro, listar formatos disponÃ­veis para debug
        try:
            info = ydl.extract_info(url, download=False)
            print("ğŸ“‹ Formatos disponÃ­veis:")
            for f in info.get('formats', []):
                if f.get('audio_ext') != 'none':
                    print(f"  â€¢ {f.get('format_id')}: {f.get('ext')} {f.get('resolution')}")
        except:
            pass

        # Tentar download
        ydl.download([url])

    # Procurar arquivo resultante
    for ext in ['.mp3', '.m4a', '.webm', '.opus', '.ogg']:
        possible_file = f"{output_file}{ext}"
        if os.path.exists(possible_file):
            print(f"âœ… Ãudio encontrado: {possible_file}")
            return os.path.abspath(possible_file)

    # Se nÃ£o encontrou com extensÃ£o, procurar qualquer arquivo que comece com output_file
    for file in os.listdir():
        if file.startswith(output_file):
            print(f"âœ… Usando arquivo: {file}")
            return os.path.abspath(file)

    raise FileNotFoundError("âŒ Nenhum arquivo de Ã¡udio foi gerado")
```
